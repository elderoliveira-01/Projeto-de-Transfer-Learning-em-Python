```python
from keras.applications import vgg16
from keras.preprocessing.image import load_img, img_to_array
from keras.models import Model
from keras.applications.imagenet_utils import preprocess_input

import os
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Caminho correto onde as imagens estão localizadas
base_path = "/content/dataset_imagens"  # Diretório base onde os subdiretórios estão localizados
subdirs = ["bracelet", "watch", "necklace", "ring"]  # Subdiretórios onde as imagens estão armazenadas
imgs_model_width, imgs_model_height = 224, 224
nb_closest_images = 5  # Número de imagens mais similares a recuperar

# Carregar o modelo pré-treinado VGG16
vgg_model = vgg16.VGG16(weights='imagenet')

# Remover as últimas camadas para extrair características em vez de previsões
feat_extractor = Model(inputs=vgg_model.input, outputs=vgg_model.get_layer("fc2").output)

# Exibir as camadas do modelo
feat_extractor.summary()

# Verifique se os diretórios existem e carregue as imagens
files = []
for subdir in subdirs:
    subdir_path = os.path.join(base_path, subdir)
    if os.path.exists(subdir_path):
        files.extend([os.path.join(subdir_path, x) for x in os.listdir(subdir_path) if x.lower().endswith(('.png', '.jpg'))])

# Verifique a quantidade de arquivos encontrados
print("Arquivos encontrados:", len(files))

# Função para carregar e exibir a imagem
def show_image(image_path, title="Imagem"):
    original = load_img(image_path, target_size=(imgs_model_width, imgs_model_height))
    plt.imshow(original)
    plt.title(title)
    plt.axis('off')  # Desliga os eixos
    plt.show()

# **Alteração aqui**: Defina a imagem original a partir de um caminho específico ou índice.
# Altere o caminho da imagem para o desejado aqui. Se quiser alterar, só trocar o caminho ou índice de files.

given_img = "/content/dataset_imagens/bracelet/Image_16.jpg"  # Caminho específico para a imagem

# **Exibição da imagem original**
# A imagem original será carregada e mostrada assim que o código for executado
show_image(given_img, "Imagem Original")

# Converter a imagem para um array NumPy
original = load_img(given_img, target_size=(224, 224))  # Garantir que a imagem seja carregada corretamente
numpy_image = img_to_array(original)

# Converter a imagem em formato de lote
image_batch = np.expand_dims(numpy_image, axis=0)
print('Tamanho do lote de imagem:', image_batch.shape)

# Pré-processar a imagem para o modelo VGG16
processed_image = preprocess_input(image_batch.copy())

# Extrair as características da imagem carregada
img_features = feat_extractor.predict(processed_image)
print("Características extraídas com sucesso!")
print("Número de características da imagem:", img_features.size)

# Carregar e processar todas as imagens no conjunto de dados
importedImages = []
for f in files:
    original = load_img(f, target_size=(224, 224))
    numpy_image = img_to_array(original)
    image_batch = np.expand_dims(numpy_image, axis=0)
    importedImages.append(image_batch)

# Empilhar todas as imagens para criar um lote
images = np.vstack(importedImages)

# Pré-processar as imagens para o modelo VGG16
processed_imgs = preprocess_input(images.copy())

# Extrair as características de todas as imagens
imgs_features = feat_extractor.predict(processed_imgs)
print("Características extraídas com sucesso para todas as imagens!")
print("Formato das características das imagens:", imgs_features.shape)

# Calcular as similaridades cossenoidais entre as imagens
cosSimilarities = cosine_similarity(imgs_features)

# Armazenar os resultados em um DataFrame do pandas
cos_similarities_df = pd.DataFrame(cosSimilarities, columns=files, index=files)
print(cos_similarities_df.head())

# Função para recuperar os produtos mais similares a uma imagem dada
def retrieve_most_similar_products(given_img):
    print("-----------------------------------------------------------------------")
    print("Produto original:")

    # Recuperar as imagens mais similares com base na similaridade cosseno
    closest_imgs = cos_similarities_df[given_img].sort_values(ascending=False)[1:nb_closest_images+1].index
    closest_imgs_scores = cos_similarities_df[given_img].sort_values(ascending=False)[1:nb_closest_images+1]

    # Exibir a imagem original junto com as imagens mais similares
    fig, axes = plt.subplots(1, nb_closest_images + 1, figsize=(15, 5))

    # Exibir a imagem original
    original_img = load_img(given_img, target_size=(imgs_model_width, imgs_model_height))
    axes[0].imshow(original_img)
    axes[0].set_title("Imagem Original")
    axes[0].axis('off')  # Desliga os eixos

    # Exibir as imagens mais similares
    for i in range(nb_closest_images):
        similar_img = load_img(closest_imgs[i], target_size=(imgs_model_width, imgs_model_height))
        axes[i + 1].imshow(similar_img)
        axes[i + 1].set_title(f"Similar {i+1}")
        axes[i + 1].axis('off')  # Desliga os eixos

    plt.show()  # Mostrar as imagens lado a lado

    # Exibir as pontuações de similaridade
    for i in range(len(closest_imgs)):
        print(f"Imagem Similar {i+1}: {closest_imgs[i]} - Similaridade: {closest_imgs_scores.iloc[i]}")

# **Aqui você escolhe qual imagem deseja usar como produto original**
# Alterando para a imagem específica
retrieve_most_similar_products(given_img)
```
